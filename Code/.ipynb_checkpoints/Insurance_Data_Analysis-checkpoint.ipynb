```python
# Import Libraries
import json
from io import StringIO
import gzip
import pandas as pd
from numpy import isnan
```

```python
#Read the JSON file
with gzip.open ('/content/drive/MyDrive/Insurance_Data/NV_CACHMED0000.json.gz','r') as json_file:
  data = json.loads(json_file.read().decode("ascii"))
```

```python
#Temprorary DataFrame to understand the JSON data and change it to dictionary
df = pd.DataFrame(dict([ (k, pd.Series(v)) for k,v in data.items()]))
```

```python
#Get Provider References with Group_id including Provider Groups with tin and NPI
provider_df={}
for op in df['provider_references']:
  provider_groups = []
  provider_type = []
  for h in op.get('provider_groups'):
    provider_type.append(h.get('tin').get('type'))
    provider_groups.append(h.get('tin').get('value'))
  provider_df.update({str(op.get('provider_group_id')):(provider_groups,provider_type)})
```

```python
#Create desired DataFrame to save data based on the Excel input from the file
xls = pd.ExcelFile('/content/drive/MyDrive/Insurance_Data/data_needed.xlsx')
ins_database = pd.read_excel(xls, xls.sheet_names[2])
# df2 = pd.read_excel(xls, 'Sheet2')
ins_database.drop("Field", axis=1,inplace=True)
ins_database.drop(ins_database.index,inplace=True)
temp_ins_df = ins_database
```

```python
# Pull data from each JSON row and update 
for u in df['in_network']:
  if pd.isnull(u) ==False:
    billing_code = u.get('billing_code')
    billing_code_type =u.get('billing_code_type')
    billing_code_type_version =u.get('billing_code_type_version')
    name =u.get('name')
    negotiation_arrangement =u.get('negotiation_arrangement')

    provider_references=[]
    billing_class=[]
    expiration_date=[]
    negotiated_rate=[]
    negotiated_type=[]
    service_code=[]
    
    for i in u.get('negotiated_rates'):
      provider_references.append(i.get('provider_references')[0])
      billing_class.append(i.get('negotiated_prices')[0].get('billing_class'))
      expiration_date.append(i.get('negotiated_prices')[0].get('expiration_date'))
      negotiated_rate.append(i.get('negotiated_prices')[0].get('negotiated_rate'))
      negotiated_type.append(i.get('negotiated_prices')[0].get('negotiated_type'))
      service_code.append(i.get('negotiated_prices')[0].get('service_code'))
    
    provider_groups_ls = []
    type_ls= []
    for onum in provider_references:
      provider_groups_ls.append(provider_df.get(str(onum))[0])
      type_ls.append(provider_df.get(str(onum))[1])

    #Update temprorary dataframe for each Bill
    temp_ins_df['billing_class'] = billing_class
    temp_ins_df['expiration_date'] = expiration_date
    temp_ins_df['negotiated_rate'] = negotiated_rate
    temp_ins_df['negotiated_type'] = negotiated_type
    temp_ins_df['service_code'] = service_code
    temp_ins_df['provider_group_id'] = provider_references
    temp_ins_df['provider_groups'] = provider_groups_ls
    temp_ins_df['npi'] = provider_groups_ls
    temp_ins_df['type'] = type_ls
    temp_ins_df['billing_code'] = billing_code
    temp_ins_df['billing_code_type'] = billing_code_type
    temp_ins_df['billing_code_type_version'] = billing_code_type_version
    temp_ins_df['name'] = name
    temp_ins_df['description'] = name
    temp_ins_df['negotiation_arrangement'] = negotiation_arrangement
    temp_ins_df['reporting_entity_name'] = df['reporting_entity_name'][0]
    temp_ins_df['reporting_entity_type'] = df['reporting_entity_type'][0]
    temp_ins_df['last_updated_on'] = df['last_updated_on'][0]
    temp_ins_df['version'] = df['version'][0]

    temp_ins_df['provider_groups.1'] = provider_groups_ls
    temp_ins_df['billing_code.1'] = billing_code
    temp_ins_df['billing_code_type.1'] = billing_class
    temp_ins_df['billing_code_type_version.1'] = billing_code_type_version
    temp_ins_df['description.1'] = name

    temp_ins_df['billing_code.2'] = billing_code
    temp_ins_df['billing_code_type.2'] = billing_class
    temp_ins_df['billing_code_type_version.2'] = billing_class
    temp_ins_df['description.2'] = name
    
    #Concatenate data for all the bills
    ins_database = pd.concat([temp_ins_df, ins_database], axis=0)
    temp_ins_df.drop(temp_ins_df.index,inplace=True)
```
